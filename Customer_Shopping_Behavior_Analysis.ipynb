```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine(
    "mysql+pymysql://root:tanveer%4030@localhost/customer_behavior"
)
```

```python
## what is the total revenue generated by males vs female customers?

q1="select gender,sum(purchase_amount) as revenue from customer_behavior group by gender"
df1=pd.read_sql(q1,engine)
df1
```

```python
## which customers used a discount but still spent mare than avg purchase amount?
q2="select customer_id,purchase_amount from customer_behavior where discount_applied='Yes' and purchase_amount>=(select avg(purchase_amount)from customer_behavior)"
df2=pd.read_sql(q2,engine)
df2
```

```python
## which are the top 5 products with the highest avg review rating ?
q3="select item_purchased, round(avg(review_rating),2) as avg_review_rating from customer_behavior group by item_purchased order by avg_review_rating desc limit 5;"
df3=pd.read_sql(q3,engine)
df3
```

```python
## compare  the avg purchase amount between standard and express shipping ?
q4="select shipping_type, round(avg(purchase_amount),2)as avg_amount from customer_behavior where shipping_type in ('Express','Standard') group by shipping_type;"
df4=pd.read_sql(q4,engine)
df4
```

```python
## do subscribed customers spend more ? 
## compare avg spent and total revenue between subscribers and non-subscribers?
q5="select subscription_status,count(customer_id) as total_customers,round(avg(purchase_amount),2) as avg_amount,round(sum(purchase_amount),2) as total_revenue from customer_behavior group by subscription_status order by total_revenue,avg_amount desc;"
df5=pd.read_sql(q5,engine)
df5
```

```python
## which 5 products have the highest percentage of purchase with discount applied?
q6="select item_purchased,round(100*sum( case when discount_applied = 'Yes' then 1 else 0 end)/count(*) * 100,2) as discount_rate from customer_behavior group by item_purchased order by discount_rate desc limit 5;"
df6=pd.read_sql(q6,engine)
df6
```

```python
## segment the customers into new,returning and loyal based on their total numbeer of previous purchases and show the count of each segment?
q7="with customer_type as(select customer_id,previous_purchases,case when previous_purchases=1 then 'New' when previous_purchases between 2 and 10 then 'Returning'else 'Loyal'end as customer_segment from customer_behavior) select customer_segment,count(*) as 'Number of customers' from customer_type group by customer_segment;"
df7=pd.read_sql(q7,engine)
df7
```

```python
## ## what are the top 3 most purchased products with in each category?
q8="with item_count as(select category,item_purchased,count(customer_id) as total_orders,row_number() over (partition by category order by count(customer_id) desc) as item_rank from customer_behavior group by category,item_purchased) select item_rank,category,item_purchased,total_orders from item_count where item_rank<=3;"
df8=pd.read_sql(q8,engine)
df8
```

```python
## are the customers who repeatbbuyers (more than 5 previous purchases) also likely to subscribe?
q9="select subscription_status,count(customer_id) as repeat_buyers from customer_behavior where previous_purchases>5 group by subscription_status;"
df9=pd.read_sql(q9,engine)
df9
```

```python
## what is the revenue contribution of each age group?
q10="select age_group,sum(purchase_amount) as total_revenue from customer_behavior group by age_group order by total_revenue desc;"
df10=pd.read_sql(q10,engine)
df10
```
